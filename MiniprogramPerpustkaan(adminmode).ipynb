```python
#List Peminjamanan Buku, Nama Peminjam dan Tanggal Peminjaman
Buku_kembali=['Sejarah Kendaraan','Sastra Mesin','Teknik Untuk Umum']

Buku_pinjam=[]

Siswa=[]

tgl_pinjam_buku=[]

password_Admin=['PerPUS M3S1N']
#Looping Opsi            
while True:
            print(''' ======PERPUSTAKAAN MESIN=======
                  1. Tampilkan semua buku yang tersedia
                  2. Tampilkan semua peminjam perpustakaan
                  3. Pinjam buku
                  4. Kembalikan buku
                  5. Daftar Peminjam Buku
                  6. Mode Admin
                  7. Keluar
                  ''')
            choice=int(input('Masukkan Pilihan:'))

            #Opsi 1(Menampilkan list buku perpustakaan)
            if choice==1:
                        print('Buku-buku di perpustakaan kami adalah sebagai berikut:')
                        print('================================')
                        for book in Buku_kembali:
                          print(book)

            #Opsi 2(Menampilkan list nama peminjam di perpustakaan)
            elif choice==2:
                        if not Siswa:
                          print('Tidak ada peminjam')
                        else:
                          print('Peminjam di Perpustakaan ini adalah:')
                          print('================================')
                          for student in Siswa:
                            print(student)

            #Opsi 3(Memulai Proses Peminjaman)
            elif choice==3:
                        pinjam_buku=input('Masukkan Judul Buku yang ingin Dipinjam>>').title()
                        print(pinjam_buku) 
                        if pinjam_buku in Buku_kembali:
                          Buku_kembali.remove(pinjam_buku)
                          Buku_pinjam.append(pinjam_buku)
                          tgl_pinjam=input('Masukkan tanggal peminjaman (YYYY-MM-DD):>>')
                          print(tgl_pinjam)
                          tgl_pinjam_buku.append(tgl_pinjam)
                          daftar_siswa=input('Masukkan nama peminjam>>').title()
                          print(daftar_siswa)
                          Siswa.append(daftar_siswa)                              
                          print('Buku Berhasil Dipinjam')
                        else:
                            print('Maaf, Buku Tidak Tersedia di Perpustakaan')

            #Opsi 4(Memulai Proses Pengembalian)
            elif choice==4:
                        kembali_buku=input('Masukkan Judul Buku yang Ingin Dikembalikan>>').title()
                        print(kembali_buku)
                        if kembali_buku in Buku_pinjam: 
                         Buku_pinjam.remove(kembali_buku)
                         Buku_kembali.append(kembali_buku)
                         tgl_kembali=input('Masukkan Tanggal Peminjaman (YYYY-MM-DD):>>')
                         print(tgl_kembali)
                         tgl_pinjam_buku.remove(tgl_kembali)
                         siswa_pinjam=input('Masukkan Nama Peminjam>>').title()
                         print(siswa_pinjam)
                         Siswa.remove(siswa_pinjam)                    
                         print('Terima Kasih, Buku Telah Dikembalikan')
                        else:
                         print('Maaf, Buku Tidak Masuk Daftar Peminjaman')
                        
            #Opsi 5(Menampilkan List data peminjaman)
            elif choice==5:
                         if not tgl_pinjam_buku:
                          if not Siswa:
                            if not Buku_pinjam:
                              print('Belum ada peminjaman')
                  
                         else:
                          print('Berikut daftar Tanggal Peminjaman, Peminjam dan Judul Buku yang Dipinjam:')
                          print('================================')
                          for a in tgl_pinjam_buku,Siswa,Buku_pinjam:
                            print(a)
            
            #Opsi 6(Mode Admin)
            elif choice==6:
                Pass_admin=input('Masukkan Password Admin>>')
                if password_Admin in Pass_admin:
                  print('''====ADMIN MODE===
                          1.Sumbangan Buku
                          2.Penghapusan Buku
                          3.Kembali ke Menu Utama''')
                  choiceAdmin=int(input('Masukkan Pilihan:'))
                              if choiceAdmin==1:
                                  tambah_buku=input('Masukkan Judul Buku yang Ingin Disumbangkan>>').title()
                                  Buku_kembali.append(tambah_buku) 
                                  print('Terima Kasih Telah Memberikan Buku', tambah_buku, ', Semoga Dapat Dimanfaatkan dengan Baik')
                              elif choiceAdmin==2:
                                Hapus_buku=input('Masukkan Judul Buku yang Ingin Dihapuskan>>').title()
                                if Hapus_buku in Buku_kembali:
                                Buku_kembali.remove(Hapus_buku)
                                print('Buku', Hapus_buku, 'Berhasil Dihapuskan')
                                else:
                                print('Maaf, Buku', Hapus_buku, 'Tidak Masuk Daftar Peminjaman')
                                elif choiceAdmin==3:
                                print('Keluar dari Mode Admin ')           
                                else:
                                print('Pilihan Tidak Tersedia Silahkan Pilih Ulang')  
                                main()
                else:
                 print('Password Salah')
                                                          
            #Opsi 7(Exit program perpustakaan)
            elif choice==7:
                  print('Terima Kasih Telah Mengunjungi Perpustakaan Mesin, Selamat Belajar :D')
                  break   
            else:
                  print('Pilihan Tidak Tersedia Silahkan Pilih Ulang')    
```
